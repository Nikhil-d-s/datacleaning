Data Cleaning

#Outlier detection in 'Age' column=IQR

```python
q1=df['Age'].quantile(0.25)
q1
```

```python
q3=df['Age'].quantile(0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
lower=q1-1.5*iqr
lower
```

```python
upper=q3+1.5*iqr
upper
```

```python
df[(df['Age']<lower) & (df['Age']>upper)]
#this is used for findling outliars
```

```python
q1=df['Salary'].quantile(0.25)
q1
```

```python
q3=df['Salary'].quantile(0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
lower=q1-1.5*iqr
lower
```

```python
upper=q3+1.5*iqr
upper
```

```python
df[(df['Salary']<lower) & (df['Salary']>upper)]
```
